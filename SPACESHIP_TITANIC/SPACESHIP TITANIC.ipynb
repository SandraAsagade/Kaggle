## FILLING THE MISSING VALUES

In this section, we fill the missing values in the data.

```python
train_data1=train_data.copy()
#AGE
train_data1['Age']=train_data1['Age'].interpolate()
train_data1['HomePlanet']=train_data1['HomePlanet'].fillna('Earth')
train_data1['Destination']=train_data1['Destination'].fillna('TRAPPIST-1e')
train_data1['Cabin']=train_data1['Cabin'].fillna('G/734/S')
train_data1['CryoSleep']=train_data1['CryoSleep'].fillna('False')
train_data1['VIP']=train_data1['VIP'].fillna('False')
train_data1['Name']=train_data1['Name'].fillna('Missing')

test_data1=test_data.copy()
test_data1['Age']=test_data1['Age'].interpolate(method='linear')
test_data1['HomePlanet']=test_data1['HomePlanet'].fillna('Earth')
test_data1['Destination']=test_data1['Destination'].fillna('TRAPPIST-1e')
test_data1['Cabin']=test_data1['Cabin'].fillna('G/160/P')
test_data1['CryoSleep']=test_data1['CryoSleep'].fillna('False')
test_data1['VIP']=test_data1['VIP'].fillna('False')
test_data1['Name']=test_data1['Name'].fillna('Missing')
```

```python
train_data1.isnull().sum()
```

## DROPPING THE NUMERICAL FEATURES EXCEPT AGE
Here we look at the data realistically to see which feature will determine if a passenger was transported or not. And from looking at the data, we can conclude that the expenditure features had no valid effect on a passenger being transported or not. So we drop the columns with these features.

```python
train_data2=train_data1.copy()
train_data2=train_data2.drop(['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck'],axis=1)

test_data2=test_data1.copy()
test_data2=test_data2.drop(['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck'],axis=1)
```

## PASSENGER ID AND CABIN FEATURES
Here we call the lambda function to get the Cabin Number, Cabin Deck, Cabin Side, Passenger Group and Passenger Number, as these features give quality predictions.

```python
train_data3=train_data2.copy()
#PASSENGER
train_data3['PassengerGroup']=train_data3['PassengerId'].apply(lambda x: x.split('_')[0]).astype(int)
train_data3['PassengerNo']=train_data3['PassengerId'].apply(lambda x: x.split('_')[1]).astype(int)
#CABIN
train_data3['CabinDeck']=train_data3['Cabin'].apply(lambda x: x.split('/')[0])
train_data3['CabinNo']=train_data3['Cabin'].apply(lambda x: x.split('/')[1]).astype(int)
train_data3['CabinSide']=train_data3['Cabin'].apply(lambda x: x.split('/')[2])


test_data3=test_data2.copy()
#PASSENGER
test_data3['PassengerGroup']=test_data3['PassengerId'].apply(lambda x: x.split('_')[0]).astype(int)
test_data3['PassengerNo']=test_data3['PassengerId'].apply(lambda x: x.split('_')[1]).astype(int)
#CABIN
test_data3['CabinDeck']=test_data3['Cabin'].apply(lambda x: x.split('/')[0])
test_data3['CabinNo']=test_data3['Cabin'].apply(lambda x: x.split('/')[1]).astype(int)
test_data3['CabinSide']=test_data3['Cabin'].apply(lambda x: x.split('/')[2])

#DROPPING PassengerId and Cabin COLUMNS

train_data3=train_data3.drop(['Cabin','Name'],axis=1)
test_data3=test_data3.drop(['Cabin','Name'],axis=1)
```

```python
train_data3
```

## WORKING ON THE AGE (BINNING)

```python
#TRAIN DATA
train_data3['AgeClass']=train_data3['Age']
train_data3.loc[train_data3['AgeClass']<=12,'AgeClass']=0 #child
train_data3.loc[(train_data3['AgeClass']>12) & (train_data3['AgeClass']<=17),'AgeClass']=1 #teenager
train_data3.loc[(train_data3['AgeClass']>17) & (train_data3['AgeClass']<=30),'AgeClass']=2 #young adults
train_data3.loc[(train_data3['AgeClass']>30) & (train_data3['AgeClass']<=49),'AgeClass']=3 #middle-aged adults
train_data3.loc[(train_data3['AgeClass']>49) & (train_data3['AgeClass']<=80),'AgeClass']=4 #old adults

train_data3['AgeGroup']=train_data3['AgeClass']
train_data3['AgeRange']=train_data3['AgeClass']
train_data3['AgeRange'].replace([0.0,1.0,2.0,3.0,4.0],['under 13yrs','(13-17)yrs','(18-30)yrs','(31-49)yrs','(50-80)yrs'],inplace=True)
train_data3['AgeGroup'].replace([0.0,1.0,2.0,3.0,4.0],['Child','Teenager','Young Adult','Mid-Aged Adult','Old-Aged Adults'],inplace=True)
train_data3=train_data3.drop(['AgeClass'],axis=1)

#TEST DATA
test_data3['AgeClass']=test_data3['Age']
test_data3.loc[test_data3['AgeClass']<=12,'AgeClass']=0 #child
test_data3.loc[(test_data3['AgeClass']>12) & (test_data3['AgeClass']<=17),'AgeClass']=1 #teenager
test_data3.loc[(test_data3['AgeClass']>17) & (test_data3['AgeClass']<=30),'AgeClass']=2 #young adults
test_data3.loc[(test_data3['AgeClass']>30) & (test_data3['AgeClass']<=49),'AgeClass']=3 #middle-aged adults
test_data3.loc[(test_data3['AgeClass']>49) & (test_data3['AgeClass']<=80),'AgeClass']=4 #old adults

test_data3['AgeGroup']=test_data3['AgeClass']
test_data3['AgeRange']=test_data3['AgeClass']
test_data3['AgeRange'].replace([0.0,1.0,2.0,3.0,4.0],['under 13yrs','(13-17)yrs','(18-30)yrs','(31-49)yrs','(50-80)yrs'],inplace=True)
test_data3['AgeGroup'].replace([0.0,1.0,2.0,3.0,4.0],['Child','Teenager','Young Adult','Mid-Aged Adult','Old-Aged Adults'],inplace=True)
test_data3=test_data3.drop(['AgeClass'],axis=1)
```

## MODELLING
We will begin this by dropping some columns

```python
train_data4=train_data3.copy()
train_data4=train_data4.drop(['PassengerId','PassengerGroup','CabinNo','AgeRange','Age'],axis=1)
test_data4=test_data3.copy()
test_data4=test_data4.drop(['PassengerId','PassengerGroup','CabinNo','AgeRange','Age'],axis=1)
```

```python
train_data4.info()
```

### LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder as LE
#TRAIN DATA
train_data5=train_data4.copy()
train_data5['HomePlanet']=LE().fit_transform(train_data5['HomePlanet'])
train_data5['CryoSleep'].replace([False,True],[0,1],inplace=True)
train_data5['Destination']=LE().fit_transform(train_data5['Destination'])
train_data5['VIP'].replace([False,True],[0,1],inplace=True)
train_data5['CabinDeck']=LE().fit_transform(train_data5['CabinDeck'])
train_data5['CabinSide']=LE().fit_transform(train_data5['CabinSide'])
train_data5['AgeGroup']=LE().fit_transform(train_data5['AgeGroup'])
train_data5['Transported'].replace([False,True],[0,1],inplace=True)

#TEST DATA
test_data5=test_data4.copy()
test_data5['HomePlanet']=LE().fit_transform(test_data5['HomePlanet'])
test_data5['CryoSleep'].replace([False,True],[0,1],inplace=True)
test_data5['Destination']=LE().fit_transform(test_data5['Destination'])
test_data5['VIP'].replace([False,True],[0,1],inplace=True)
test_data5['CabinDeck']=LE().fit_transform(test_data5['CabinDeck'])
test_data5['CabinSide']=LE().fit_transform(test_data5['CabinSide'])
test_data5['AgeGroup']=LE().fit_transform(test_data5['HomePlanet'])
```

### CONVERTING OBJECT TO FLOAT
I really do not know why after label encoding of some features I got an error, so here we will convert strings/bool to float to terminate that error.

```python
train_data5['CryoSleep']=pd.to_numeric(train_data5['CryoSleep'],errors='coerce')
train_data5['VIP']=pd.to_numeric(train_data5['VIP'],errors='coerce')
train_data5.fillna(0,inplace=True)

test_data5['CryoSleep']=pd.to_numeric(test_data5['CryoSleep'],errors='coerce')
test_data5['VIP']=pd.to_numeric(test_data5['VIP'],errors='coerce')
test_data5.fillna(0,inplace=True)
```

### X_TRAIN,X_TEST,Y_TRAIN,Y_TEST
The data has already been splitted into the train and test data. What is left is for us to assign them to their respective x_train,y_train,x_test and y_test. Oh! and for y_test, we will have to predict that as we only have x_test from the test data.

```python
x_train=train_data5.iloc[:,[0,1,2,3,5,6,7,8]].values
y_train=train_data5.iloc[:,4].values
x_test=test_data5.iloc[:,[0,1,2,3,4,5,6,7]].values
```

```python
y_train
```

```python
x_test
```

### MODEL ALGORITHMS
Here we will try out different classification models and see which one predicts best.

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression(C=0.01, penalty= 'l2',random_state=22).fit(x_train,y_train)
accuracy_score1=model1.score(x_train,y_train)
prediction1=model1.predict(x_test)

#DECISION TREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier(criterion= 'gini', max_depth= 5, min_samples_leaf= 2, min_samples_split= 3, splitter= 'best',random_state=22).fit(x_train,y_train)
accuracy_score2=model2.score(x_train,y_train)
prediction2=model2.predict(x_test)

#RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier(n_estimators=100,criterion='entropy',random_state=22).fit(x_train,y_train)
accuracy_score3=model3.score(x_train,y_train)
prediction3=model3.predict(x_test)

#SVC
from sklearn.svm import SVC
model4=SVC(C= 100, kernel= 'poly',degree=2,random_state=22).fit(x_train,y_train)
accuracy_score4=model4.score(x_train,y_train)
prediction4=model4.predict(x_test)

#GRADIENT BOOSTING CLASSIFIER
from sklearn.ensemble import GradientBoostingClassifier
model5=GradientBoostingClassifier().fit(x_train,y_train)
accuracy_score5=model5.score(x_train,y_train)
prediction5=model5.predict(x_test)

#XGBCLASSIFIER
from xgboost import XGBClassifier
model6=XGBClassifier(eval_metric='logloss').fit(x_train,y_train)
accuracy_score6=model6.score(x_train,y_train)
prediction6=model6.predict(x_test)
```

## HEATMAP
This will check the number of times the models predicted the train data correctly, where True = 1 and False = 0

```python
from sklearn.metrics import confusion_matrix

pred1=model1.predict(x_train)
model1_cm=confusion_matrix(y_train,pred1)
pred2=model2.predict(x_train)
model2_cm=confusion_matrix(y_train,pred2)
pred3=model3.predict(x_train)
model3_cm=confusion_matrix(y_train,pred3)
pred4=model4.predict(x_train)
model4_cm=confusion_matrix(y_train,pred4)
pred5=model5.predict(x_train)
model5_cm=confusion_matrix(y_train,pred5)
pred6=model6.predict(x_train)
model6_cm=confusion_matrix(y_train,pred6)

#HEATMAP PLOT
pp.figure(figsize=(20,10))

pp.subplot(2,4,1)
pp.title("LogisticRegression_cm")
sb.heatmap(model1_cm,annot=True,cmap="Blues",fmt="d",cbar=False)

pp.subplot(2,4,2)
pp.title("DecisionTree_cm")
sb.heatmap(model2_cm,annot=True,cmap="Blues",fmt="d",cbar=False)

pp.subplot(2,4,3)
pp.title("RandomForest_cm")
sb.heatmap(model3_cm,annot=True,cmap="Oranges",fmt="d",cbar=False)

pp.subplot(2,4,4)
pp.title("SVC_cm")
sb.heatmap(model4_cm,annot=True,cmap="Blues",fmt="d",cbar=False)

pp.subplot(2,4,5)
pp.title("GradientBoosting_cm")
sb.heatmap(model5_cm,annot=True,cmap="Reds",fmt="d",cbar=False)

pp.subplot(2,4,6)
pp.title("XGBClassifier_cm")
sb.heatmap(model6_cm,annot=True,cmap="Blues",fmt="d",cbar=False)
```
