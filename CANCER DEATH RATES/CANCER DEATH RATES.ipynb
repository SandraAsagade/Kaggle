## TOTAL NUMBER OF CANCER DEATHS FOR A PARTICULAR COUNTRY FROM 1990-2019

```python
Cancer_DR['Total Number of Deaths']=Cancer_DR.iloc[:,3:32].sum(axis=1)#iloc function locates rows and features by their indexes (';' specifies all rows, while '3:32' specifies column 3 to column 32. when counting columns, we always start count from zero(0))
Cancer_DR
```

So now, we have added a new column added to the Cancer Death DataFrame which shows the total number of deaths for a country/region, in a particular year

## CANCER DEATH RATE IN THE WORLD FROM 1990-2019

```python
WRLD=Cancer_DR[Cancer_DR['Entity']=='World']#DataFrame of all te rows and columns about 'World' Entity
WRLD_Death=WRLD['Total Number of Deaths']#DataFrame of all te rows and columns about 'Total Number of Deaths' Entity
WRLD_Year=WRLD['Year']
WRLD_Data=pd.concat([WRLD_Year,WRLD_Death],axis=1)#This joins two DataFrames together, column by column(axis=1)
WRLD_Data.reset_index(inplace=True)#This resets the index of the concated DataFrames
#WRLD_Data
```

```python
plt.figure(figsize=(15,8))#This is the size you want your plot to have

sb.barplot(WRLD_Year,WRLD_Death,palette='Blues_d')
plt.title('Cancer Death Rate in the World From 1990 - 2019',pad=float(20),fontdict={'fontsize':float(20),'fontweight':int(2)})
plt.xlabel('Year',labelpad=float(20),fontdict={'fontsize':float(15),'fontweight':int(2)})
plt.ylabel('Death Rate',labelpad=float(20),fontdict={'fontsize':float(15),'fontweight':int(2)})
```

## TABULAR REPRESENTATION OF THE ABOVE BAR CHART

```python
percentage=[]
for i in WRLD_Data['Total Number of Deaths']:#For each value(i) in the column 'Total Number of Deaths'
    pct=i/WRLD_Data['Total Number of Deaths'].sum(axis=0) * 100 #That value(i)/sum of all the values in the column 'Total Number of Deaths'
    percentage.append(round(pct,2))#Adds pct to the list 'percentage=[]' and round it up to 2 Decimal places
WRLD_Data['Death %']=percentage#adds a new column 'Percentage' to the data frame

WRLD_Data.drop('index',axis=1,inplace=True)
WRLD_Data
```

The above bar chart shows the rate at which the death rate increased from 1990 - 2019, and from the tabular representation, the death rate increased by 1.1% - 1.15% each year from 1990 - 2019
